## Part C:- Data understanding and cleaning

```python
df["age"]
```

```python
df["age"] = pd.to_numeric(df["age"], errors="coerce")
df["age"] = df["age"].fillna(df["age"].mean()).astype(int)
```

```python
df["gender"]
```

```python
df["gender"] = df["gender"].str.upper().replace({
    "M":"MALE",
    "F":"FEMALE"
})

df["gender"] = df["gender"].str.upper().replace({
    "MALE": 0,
    "FEMALE": 1
})
```

```python
df["gender"]
```

```python
df["email"]
```

```python
df["email"] = df["email"].fillna("Unknown")
```

```python
df["income"]
```

```python
df["income"] = pd.to_numeric(df["income"], errors = "coerce")
df["income"] = df["income"].replace("not available", np.nan).fillna(df["income"].mean())
```

```python
df["signup_date"] = pd.to_datetime(df["signup_date"], errors = "coerce").fillna(method = "bfill")
```

```python
df["purchase_amount"]
```

```python
df["purchase_amount"] = pd.to_numeric(df["purchase_amount"], errors = "coerce")
df["purchase_amount"] = df["purchase_amount"].fillna(df["purchase_amount"].mean()).astype(int)
```

```python
df["purchase_amount"] 
```

```python
df["churn"] = df["churn"].str.upper()
```

## Part D:- Exploratory Data Analysis(EDA)

##  perform Univariate Analysis
### Distribution plots of Age, Income, purchase_amount

```python
plt.figure(figsize = (10,8))

plt.subplot(2,3,1)
sns.histplot(df["age"], bins = 5, kde = True)
plt.title("Age distribution")

plt.subplot(2,3,2)
sns.histplot(df["income"], kde = True)
plt.title("Income distribution")

plt.subplot(2,3,3)
sns.histplot(df["purchase_amount"], kde = True)
plt.title("Purchase_amount distribution")

plt.subplot(2,3,4)
sns.boxplot(df["age"])
plt.title("Age Boxplot")

plt.subplot(2,3,5)
sns.boxplot(df["income"])
plt.title("Income Boxplot")

plt.subplot(2,3,6)
sns.boxplot(df["purchase_amount"])
plt.title("Purchase_amount Boxplot")

plt.tight_layout()
plt.show()
```

## Perform Bivariate analysis
### Relationship between Gender and Purchase

```python
plt.figure(figsize = (10,5))

plt.subplot(1,2,1)
sns.boxplot(x="gender", y="purchase_amount", data=df)
plt.title("Gender vs Purchase Amount")

plt.subplot(1,2,2)
sns.barplot(x="gender", y="purchase_amount", data=df)
plt.title("Average Purchase by Gender")

plt.tight_layout()
plt.show()
```

### Relationship between Income and Churn

```python
plt.figure(figsize = (10,5))

plt.subplot(1,2,1)
sns.boxplot(x = "churn", y = "income", data = df)
plt.title("income vs churn amount")

plt.subplot(1,2,2)
sns.barplot(data = df, x = "churn", y = "income")

plt.tight_layout()
plt.show()
```
